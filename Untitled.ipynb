```python
def myfunc(st):
    word = ""
    for index,letter in enumerate(st):
        print(f'{index} {letter}')
        if index%2 == 0:
            print(f'even {letter.upper()}')
            
            word += letter.upper()
        else:
            print('odd letter')
            word += letter.lower()
            
    print(word)

```
